```python
#XGboost
#protein solubility prediction
#continous values of solubility
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import r2_score
from sklearn.metrics import classification_report
import random
from sklearn.metrics import accuracy_score
train = pd.read_csv('cleandata1.csv',index_col=0)
train.head()
```

```python
from sklearn.model_selection import train_test_split
#train.iloc[ : ,20] = [1 if x>=0.44 else 0 for x in train.iloc[ : ,20]]
X = train.iloc[:,:-1]
y = train.iloc[:,-1]
        
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

fname = X_train.columns
dtrain = xgb.DMatrix(X_train, y_train, feature_names=fname)
dvalid = xgb.DMatrix(X_test, y_test, feature_names=fname)

xgb_params = {
                'eta':0.05,
                'max_depth':5,
                'sub_sample':0.8,
                'colsample_bytree':0.5,
                'objective':'binary:logistic',
                'eval_metric':'auc',
                'booster': 'gbtree',
                'min_child_weight':10
            }

model = xgb.train(xgb_params, dtrain, num_boost_round=301, evals=[(dvalid, 'test')], early_stopping_rounds=50, verbose_eval=100)
best_round = model.best_iteration
print(best_round)
final_model = xgb.train((xgb_params), dtrain, num_boost_round=best_round)
dtest_final = xgb.DMatrix(X_test, y_test, feature_names=fname)
y_pred_final = final_model.predict(dtest_final)


y_pred = y_pred_final
y_true = y_test 
print(r2_score(y_true, y_pred)) 

y_pred_final = [1 if x>=0.39 else 0 for x in y_pred_final]
y_test  = [1 if x>=0.39 else 0 for x in y_test]
y_pred = y_pred_final
y_true = y_test 
accuracy_score(y_true, y_pred)
```
